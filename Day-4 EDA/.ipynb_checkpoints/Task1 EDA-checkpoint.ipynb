# Employee Salary Analysis & Visualization

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('employee_data.csv')
print(df)
```

```python
df
```

```python
# 2. Find the average salary in each department
average_salary = df.groupby('department')['salary'].mean()
print("Average Salary in Each Department:")
print(average_salary)
print("\n")
```

```python
# 3. Find the highest and lowest salary in the company
highest_salary = df['salary'].max()
lowest_salary = df['salary'].min()
print(f"Highest Salary in the Company: {highest_salary}")
print(f"Lowest Salary in the Company: {lowest_salary}")
```

```python
# 4. Find the employee with the maximum salary
employee_max_salary = df[df['salary'] == highest_salary]
print("Employee with the Maximum Salary:")
print(employee_max_salary)
```

```python
# 5. Count the number of employees in each department
employee_count = df['department'].value_counts()
print("Number of Employees in Each Department:")
print(employee_count)
```

```python
# 6. Plot a bar chart showing the number of employees in each department
plt.figure(figsize=(10, 6))
employee_count.plot(kind='bar', color='skyblue')
plt.title('Number of Employees in Each Department')
plt.xlabel('Department')
plt.ylabel('Number of Employees')
plt.xticks(rotation=45)
plt.grid(axis='y')

# Save the plot as a PNG file
plt.savefig('employee_count_by_department.png')

# Display the plot
plt.show()
```

```python
# 7. Plot a pie chart showing salary distribution by department
plt.figure(figsize=(8, 8))
df.groupby('department')['salary'].sum().plot(kind='pie', autopct='%1.1f%%', startangle=140)
plt.title('Salary Distribution by Department')
plt.ylabel('')

# Save the plot as a PNG file
plt.savefig('salary_distribution_by_department.png')

# Display the plot
plt.show()
```

```python
# 8. Scatter plot showing salary vs experience
# Assuming 'experience' column exists in the dataset
if 'experience' in df.columns:
    plt.figure(figsize=(10, 6))
    plt.scatter(df['experience'], df['salary'], color='skyblue')
    plt.title('Salary vs Experience')
    plt.xlabel('Experience (Years)')
    plt.ylabel('Salary')
    plt.grid(True)

    # Save the plot as a PNG file
    plt.savefig('salary_vs_experience.png')

    # Display the plot
    plt.show()
else:
    print("The 'experience' column is not present in the dataset.")
```

```python
# 9. Find employees who earn more than the average salary
average_salary_value = df['salary'].mean()
employees_above_average_salary = df[df['salary'] > average_salary_value]
print("Employees Who Earn More Than the Average Salary:")
print(employees_above_average_salary)
```

```python
# 10. Sort employees based on salary in descending order
sorted_employees = df.sort_values(by='salary', ascending=False)
print("Employees Sorted Based on Salary in Descending Order:")
print(sorted_employees)
```
